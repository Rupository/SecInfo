```python
from rouge_score import rouge_scorer

scorer = rouge_scorer.RougeScorer(['rougeL'], use_stemmer=True)

def calculate_rouge_avg(references, candidates):
    precision_list = []
    recall_list = []
    f1_list = []
    
    for ref, cand in zip(references, candidates):
        scores = scorer.score(ref, cand)
        precision_list.append(scores['rougeL'].precision)
        recall_list.append(scores['rougeL'].recall)
        f1_list.append(scores['rougeL'].fmeasure)
        
    avg_p = sum(precision_list) / len(precision_list)
    avg_r = sum(recall_list) / len(recall_list)
    avg_f1 = sum(f1_list) / len(f1_list)
    
    return avg_p, avg_r, avg_f1

p, r, f1 = calculate_rouge_avg(rag_tasks_eng, gemmma_rag_eng)
print(f"Gemma ENG RAG: P={p:.4f}, R={r:.4f}, F1={f1:.4f}")

p, r, f1 = calculate_rouge_avg(bias_tasks_eng, gemma_bias_eng)
print(f"Gemma ENG Bias: P={p:.4f}, R={r:.4f}, F1={f1:.4f}")

p, r, f1 = calculate_rouge_avg(rag_tasks_eng, qwen_rag_eng)
print(f"Qwen ENG RAG: P={p:.4f}, R={r:.4f}, F1={f1:.4f}")

p, r, f1 = calculate_rouge_avg(bias_tasks_eng, qwen_bias_eng)
print(f"Qwen ENG Bias: P={p:.4f}, R={r:.4f}, F1={f1:.4f}")
```
